## Conceptos b√°sicos

### Definiciones en Instrumentaci√≥n
**Sistema:**  
Un sistema puede ser descrito como un conjunto de componentes dentro de un l√≠mite que trabajan de manera conjunta para generar una salida a partir de una entrada o entradas espec√≠ficas. El l√≠mite separa al sistema de su entorno, y el sistema interact√∫a con el entorno mediante se√±ales que atraviesan dicho l√≠mite, ya sea en forma de entradas que van del entorno al sistema, o salidas que van del sistema hacia el entorno.

<img src="IMG/sistema.png" alt="sistema" width="400">

**Sistema de instrumentaci√≥n:**  
Un sistema de instrumentaci√≥n destinado a realizar mediciones recibe como entrada el valor real de la variable que se mide y genera como salida el valor que ha sido medido.

<img src="IMG/sistema_medicion.png" alt="sistema_medicion" width="400">

**Sensor `(definici√≥n 1`):**  
El sensor es el componente del sistema que est√° en contacto directo con el proceso en el que se mide una variable. Su salida depende de alg√∫n modo del valor de dicha variable, y esa salida es utilizada por el resto del sistema de medici√≥n para asignarle un valor. Un ejemplo de esto es el termopar, que tiene como entrada la temperatura y como salida una se√±al el√©ctrica peque√±a.

**Sensor `(definici√≥n 2`):**  
Un sensor es un dispositivo que transforma un fen√≥meno f√≠sico en una se√±al el√©ctrica. Los sensores son parte de la interfaz entre el mundo f√≠sico y el de los dispositivos el√©ctricos, como las computadoras. La otra parte de esta interfaz est√° formada por los actuadores, que convierten las se√±ales el√©ctricas en fen√≥menos f√≠sicos.

**Procesador de se√±al:**  
Este componente recibe la se√±al de salida del sensor y la convierte en una forma adecuada para su visualizaci√≥n o para ser transmitida a un sistema de control. Por ejemplo, en el caso de un termopar, un amplificador puede aumentar la fuerza electromotriz (e.m.f.) para que sea lo suficientemente fuerte como para ser registrada en un medidor.

**Funci√≥n de trasnferencia**

 La funci√≥n de transferencia describe c√≥mo se relacionan la se√±al de entrada f√≠sica y la se√±al de salida el√©ctrica. Normalmente, esta relaci√≥n se ilustra mediante un gr√°fico que muestra c√≥mo se conectan ambas se√±ales, y los detalles de esta conexi√≥n pueden ofrecer una descripci√≥n integral de las propiedades del sensor. En el caso de sensores caros que pasan por un proceso de calibraci√≥n individual, este v√≠nculo puede expresarse mediante una curva de calibraci√≥n certificada.

 
**Precisi√≥n:**

La precisi√≥n se refiere a la cercan√≠a de los valores medidos entre s√≠. asociado a la precisi√≥n, la incertidumbre se entiende como el mayor error esperado entre la se√±al de salida real y la ideal. A veces, esta se presenta como un porcentaje de la salida a escala completa (FSO) o como una fracci√≥n de la lectura. Por ejemplo, un term√≥metro podr√≠a garantizar una exactitud dentro del 5% de la salida m√°xima (FSO). La "precisi√≥n" es considerada un t√©rmino cualitativo en metrolog√≠a, mientras que la "incertidumbre" es un t√©rmino cuantitativo. Por ejemplo, un sensor puede tener mejor precisi√≥n que otro si su incertidumbre es del 1%, mientras que el otro tiene una incertidumbre del 3%.


**Error:**

El error es la diferencia entre el valor medido y el valor verdadero `(o el valor de referencia`):

$$
\text{Error} = \text{Valor Medido} - \text{Valor Verdadero}
$$

**Error por Histeresis:**

El error por hist√©resis es la diferencia en la salida del sistema cuando el proceso sigue un ciclo de ida y vuelta, debido a la dependencia de la direcci√≥n de la medici√≥n.  Existen sensores que no generan el mismo valor de salida cuando el est√≠mulo de entrada se aumenta o disminuye. El rango del error esperado, en relaci√≥n con la magnitud medida, se conoce como histeresis.

**Sensibilidad:**

 La sensibilidad se describe en funci√≥n de la relaci√≥n entre la se√±al f√≠sica de entrada y la se√±al el√©ctrica de salida. Usualmente, es el cociente entre una peque√±a variaci√≥n en la se√±al el√©ctrica y una peque√±a variaci√≥n en la se√±al f√≠sica. Esto puede expresarse como la derivada de la funci√≥n de transferencia respecto a la se√±al f√≠sica. Las unidades comunes son voltios/kelvin, milivoltios/kilopascal, entre otras. Un term√≥metro tendr√≠a "alta sensibilidad" si un leve cambio en la temperatura produce un gran cambio en el voltaje.

$$
\text{Sensibilidad} = \frac{\Delta \text{Salida}}{\Delta \text{Entrada}}
$$

**Span o Rango din√°mico:**

El rango de se√±ales f√≠sicas de entrada que el sensor puede convertir en se√±ales el√©ctricas se denomina rango din√°mico o intervalo. Las se√±ales que se encuentren fuera de este rango se espera que causen una inexactitud demasiado grande. Este intervalo o rango din√°mico generalmente lo proporciona el fabricante del sensor, indicando el rango en el que se espera que otras caracter√≠sticas de rendimiento descritas en las hojas de especificaciones sean v√°lidas. Las unidades comunes son kelvin, pascal, newtons, entre otras.

$$
\text{Span} = \text{Valor M√°ximo Medible} - \text{Valor M√≠nimo Medible}
$$

**No Linealidad:**

La m√°xima desviaci√≥n de una funci√≥n de transferencia lineal dentro del rango din√°mico especificado. Existen diferentes formas de medir este error, pero la m√°s com√∫n es comparar la funci√≥n de transferencia real con la ‚Äúmejor l√≠nea recta‚Äù, que est√° ubicada en el medio de las dos l√≠neas paralelas que rodean toda la funci√≥n de transferencia dentro del rango din√°mico del dispositivo. Este m√©todo es popular porque generalmente hace que los sensores muestren mejores resultados. No obstante, se pueden emplear otras l√≠neas de referencia, por lo que es importante que el usuario asegure que las comparaciones se realicen con la misma referencia. La no linealidad se refiere a la desviaci√≥n del comportamiento idealmente lineal del sensor. Una medida com√∫n de no linealidad es el error m√°ximo entre la salida del sensor y la recta ideal:

$$
\text{No Linealidad} = \frac{\text{M√°ximo Desv√≠o}}{\text{Span}} \times 100
$$

**Ruido:**

Todos los sensores generan algo de ruido adicional a la se√±al de salida. En algunos casos, el ruido del sensor es tan bajo que no afecta al sistema, ya que es menor que el ruido del siguiente componente electr√≥nico o las fluctuaciones de la se√±al f√≠sica. No obstante, en muchos casos el ruido del sensor limita el desempe√±o del sistema que lo utiliza. El ruido se distribuye generalmente a trav√©s de todo el espectro de frecuencias. Muchas fuentes comunes de ruido producen una distribuci√≥n de ruido blanco, lo que significa que la densidad espectral de ruido es constante a todas las frecuencias. Un ejemplo claro de esto es el ruido de Johnson en un resistor. Para el ruido blanco, la densidad espectral de ruido se mide en voltios/ra√≠z (Hz). Este tipo de distribuci√≥n agrega ruido a la medici√≥n con una amplitud que depende de la ra√≠z cuadrada del ancho de banda de la medici√≥n. Dado que existe una relaci√≥n inversa entre el ancho de banda y el tiempo de medici√≥n, se puede afirmar que el ruido disminuye conforme aumenta el tiempo de medici√≥n, siguiendo la ra√≠z cuadrada.

$$
\text{Ruido} = \sigma
$$

donde $\sigma$ es la desviaci√≥n est√°ndar de la se√±al de salida.

**Resoluci√≥n:**

 La resoluci√≥n de un sensor se refiere a la fluctuaci√≥n m√≠nima de la se√±al que se puede detectar. Como las fluctuaciones son fen√≥menos que ocurren en el tiempo, existe una relaci√≥n entre la duraci√≥n de la fluctuaci√≥n y la amplitud m√≠nima que se puede detectar. Por lo tanto, la definici√≥n de resoluci√≥n debe considerar el tipo de medici√≥n que se realiza. Muchos sensores tienen limitaciones debido al ruido con una distribuci√≥n espectral blanca. En estos casos, la resoluci√≥n se puede especificar en unidades de se√±al f√≠sica/ra√≠z (Hz). La resoluci√≥n real para una medici√≥n espec√≠fica se obtiene multiplicando esta cantidad por la ra√≠z cuadrada del ancho de banda de medici√≥n. Las hojas de especificaciones de los sensores suelen indicar la resoluci√≥n en unidades de se√±al/ra√≠z (Hz) o proporcionar la se√±al m√≠nima detectable para una medici√≥n concreta. Si se especifica tambi√©n la forma de la distribuci√≥n del ruido, es posible aplicar estos resultados a cualquier tipo de medici√≥n. La resoluci√≥n se refiere a la capacidad del sensor para distinguir cambios peque√±os en la variable medida. Generalmente, la resoluci√≥n es el menor cambio detectable, y se puede expresar como:

$$
\text{Resoluci√≥n} = \frac{\text{Span}}{2^N}
$$

donde \(N\) es el n√∫mero de bits de la conversi√≥n anal√≥gica a digital (ADC).

**Ancho de Banda:**

Todos los sensores presentan tiempos de respuesta limitados ante un cambio instant√°neo en la se√±al f√≠sica. Adem√°s, muchos sensores tienen tiempos de decaimiento, los cuales indican el tiempo que le toma al sensor recuperar su valor original tras un cambio abrupto en la se√±al f√≠sica. Los rec√≠procos de estos tiempos se corresponden con las frecuencias de corte superior e inferior. El ancho de banda de un sensor es el intervalo de frecuencias comprendido entre estas dos frecuencias de corte. El ancho de banda es la frecuencia m√°xima a la que el sistema puede responder sin atenuaci√≥n significativa de la se√±al. Generalmente se expresa en hertzios (Hz):

$$
\text{Ancho de Banda} = \text{Frecuencia m√°xima de respuesta}
$$

```python
import numpy as np
import plotly.graph_objects as go

# Funci√≥n de transferencia con sensibilidad ajustada por temperatura
def calcular_voltaje(acc, temp=20):
    sensibilidad_nominal = 0.167e-3  # 167 mV/g
    delta_temp = temp - 20  # Cambio de temperatura respecto a 20¬∞C
    sensibilidad_temperatura = sensibilidad_nominal * (1 + 0.00025 * delta_temp)  # Ajuste de sensibilidad
    return 1.5 + (acc * sensibilidad_temperatura)  # Calcular el voltaje

# Crear valores de aceleraci√≥n
acc_values = np.linspace(-650, 650, 1000)  # Aceleraci√≥n desde -650g a 650g

# Simular ruido con distribuci√≥n normal (media = 0, desviaci√≥n est√°ndar = 5)
ruido = np.random.normal(0, 5, acc_values.shape)

# Aplicar el ruido a los valores de aceleraci√≥n
acc_values_ruido = acc_values + ruido

# Calcular los voltajes correspondientes con aceleraci√≥n con ruido
volt_values = calcular_voltaje(acc_values_ruido)

# L√≠neas para representar la desviaci√≥n m√°xima de linealidad (¬±0.2% de la salida m√°xima)
max_voltaje = 3.5
deviation = max_voltaje * 0.002  # 0.2% de 3.5V

# La l√≠nea ideal de la funci√≥n de transferencia sin ruido
volt_lineal_ideal = calcular_voltaje(acc_values)

# L√≠neas de desviaci√≥n de linealidad (¬±0.2%)
volt_lineal_ideal_plus = volt_lineal_ideal + deviation
volt_lineal_ideal_minus = volt_lineal_ideal - deviation

# Crear la figura interactiva
fig = go.Figure()

# Agregar la gr√°fica de la funci√≥n de transferencia sin ruido (azul discontinua)
fig.add_trace(go.Scatter(x=acc_values, y=volt_lineal_ideal,
                         mode='lines', name='V = 1.5 + Acc √ó 0.167e-3 V/g (sin ruido)', 
                         line=dict(dash='dash', color='blue', width=4)))  # Hice la l√≠nea m√°s gruesa

# Agregar la gr√°fica con ruido (roja)
fig.add_trace(go.Scatter(x=acc_values, y=volt_values,
                         mode='lines', name='V con Ruido', line=dict(color='red')))

# Marcar el rango din√°mico garantizado de ¬±2g con l√≠neas verticales
fig.add_vline(x=-2, line=dict(color="green", width=2), name="Rango Garantizado (-2g a 2g)")
fig.add_vline(x=2, line=dict(color="green", width=2))

# L√≠neas de desviaci√≥n de linealidad (¬±0.2%)
fig.add_trace(go.Scatter(x=acc_values, y=volt_lineal_ideal_plus,
                         mode='lines', name='+0.2% Linealidad', line=dict(dash='dot', color='purple')))
fig.add_trace(go.Scatter(x=acc_values, y=volt_lineal_ideal_minus,
                         mode='lines', name='-0.2% Linealidad', line=dict(dash='dot', color='purple')))

# Considerar diferentes rangos de temperatura y ajustar la salida del voltaje
temperaturas = np.linspace(-10, 50, 5)  # Rango de temperaturas entre -10¬∞C y 50¬∞C
for temp in temperaturas:
    volt_temp = calcular_voltaje(acc_values, temp)  # Calcular voltaje para cada temperatura
    fig.add_trace(go.Scatter(x=acc_values, y=volt_temp,
                             mode='lines', name=f'Temperatura: {temp}¬∞C', 
                             line=dict(dash='solid', width=2)))  # Cerr√© el par√©ntesis aqu√≠

# Configurar el dise√±o de la gr√°fica
fig.update_layout(
    title="Funci√≥n de Transferencia: Aceleraci√≥n vs. Voltaje con Ruido, Efecto de Linealidad y Temperatura",
    xaxis_title="Aceleraci√≥n (g)",
    yaxis_title="Voltaje (V)",
    legend_title="Leyenda",
    showlegend=True,
    template="plotly_dark",
    xaxis=dict(range=[-650, 650]),
    yaxis=dict(range=[1.0, 3.5]),
    updatemenus=[dict(
        type="buttons",
        showactive=False,
        buttons=[dict(
            label="Restablecer Zoom",
            method="relayout",
            args=[{"xaxis.range": [-650, 650], "yaxis.range": [1.0, 3.5]}]
        )]
    )]
)

# Mostrar la gr√°fica interactiva
fig.show()

```

### Aceleraci√≥n vs. Voltaje con Ruido, Efecto de Linealidad y Temperatura  

1. **Funci√≥n de Transferencia (Sin Ruido)**  

    La l√≠nea azul discontinua representa la funci√≥n de transferencia nominal del sensor sin ning√∫n tipo de ruido.  
    Esta es la relaci√≥n ideal entre la aceleraci√≥n (en g) y el voltaje de salida (en V), dada por la ecuaci√≥n:  

    $ V = 1.5 + (Acc \times 0.167 \times 10^{-3}) $  

    Donde:  
    - $ V $ es el voltaje de salida (en V).  
    - $ Acc $ es la aceleraci√≥n (en g).  
    - $ 1.5V $ es el voltaje base cuando no hay aceleraci√≥n ($ Acc = 0 $).  
    - La sensibilidad del sensor es $ 167 $ mV/g $ = 0.167 \times 10^{-3} $ V/g.  

2. **Voltaje con Ruido**  

    La l√≠nea roja muestra el voltaje calculado con ruido.  
    El ruido se simula utilizando una distribuci√≥n normal con una desviaci√≥n est√°ndar de **5**, lo que introduce fluctuaciones en los valores de aceleraci√≥n, afectando el voltaje de salida.  
    Esta representaci√≥n nos permite ver c√≥mo las mediciones pueden variar debido a interferencias externas o incertidumbres en el sistema.  

3. **Rango Din√°mico Garantizado**  

    Las l√≠neas verdes verticales marcan el **rango din√°mico garantizado** del sensor, que es de **¬±2g**.  
    Este es el rango de aceleraci√≥n dentro del cual el sensor proporciona mediciones confiables.  
    Fuera de este rango, las mediciones podr√≠an volverse inexactas o no lineales.  

4. **Efecto de Linealidad (¬±0.2%)**  

    Las l√≠neas moradas discontinuas representan el **rango de desviaci√≥n de linealidad**, es decir, las variaciones m√°ximas permitidas respecto a la l√≠nea ideal.  
    En este caso, la **linealidad del sensor est√° garantizada dentro de un 0.2% de la salida m√°xima**.  
    Esto significa que las mediciones de voltaje pueden desviarse de la funci√≥n ideal hasta un **0.2%** debido a la naturaleza no perfecta del sensor.  

    La ecuaci√≥n para calcular la desviaci√≥n de linealidad es:  

    Desviaci√≥n Linealidad = $\pm (0.002 \times V_{max})$  

    En este caso, el voltaje m√°ximo es **3.5V**, por lo que la desviaci√≥n es:  

    Desviaci√≥n = $3.5V \times 0.002 = 0.007V = 7$  

    Por lo tanto, las l√≠neas moradas representan un desplazamiento de **¬±7 mV** respecto a la funci√≥n ideal de transferencia.  

5. **Ajuste de Sensibilidad por Temperatura**  

    Las l√≠neas adicionales con diferentes colores muestran c√≥mo el voltaje cambia con la **temperatura**.  
    El ajuste de sensibilidad por temperatura se calcula considerando un cambio de **¬±0.025% por ¬∞C** respecto a la temperatura nominal de **20¬∞C**.  

    La ecuaci√≥n de sensibilidad ajustada por temperatura es:  

    Sensibilidad Ajustada =  Sensibilidad Nominal $\times (1 + (0.00025 \times \Delta T)) $  

    Donde:  
    - **Sensibilidad Nominal** = $0.167 \times 10^{-3}$ V/g (sensibilidad a temperatura ambiente de **20¬∞C**).  
    - **$\Delta T$** es la diferencia de temperatura con respecto a **20¬∞C**.  

    Esto implica que por cada **grado Celsius** que la temperatura se desv√≠e de **20¬∞C**, la sensibilidad del sensor se ajustar√° en un **0.025%** de su valor nominal.  
    Las l√≠neas s√≥lidas en la gr√°fica representan este ajuste de sensibilidad para distintas temperaturas.  

    **Ejemplo**:  
    Si la temperatura aumenta a **30¬∞C** ($\Delta T = 10¬∞C$), la sensibilidad ajustada ser√°:  

    Sensibilidad Ajustada = $0.167 \times 10^{-3} \times (1 + (0.00025 \times 10))$  

    $=0.167 \times 10^{-3} \times 1.0025 = 0.167417 \times 10^{-3}$ V/g

### Norma para Simbolog√≠a P&ID

La simbolog√≠a utilizada en los diagramas de tuber√≠as e instrumentaci√≥n (P&ID) se puede consultar en varias normas, dependiendo de la regi√≥n y el √°mbito industrial. A continuaci√≥n, se mencionan algunas de las normas m√°s comunes:

1. **ISA-5.1**  
   - **T√≠tulo**: "Instrumentation Symbols and Identification"  
   Esta es la norma m√°s ampliamente utilizada para la simbolog√≠a de instrumentos y se√±ales en los diagramas de control, incluidos los P&ID. Define s√≠mbolos est√°ndar para representar equipos, l√≠neas de proceso, instrumentos y componentes del sistema de control.
