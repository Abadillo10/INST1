## Conceptos básicos

### Definiciones en Instrumentación
**Sistema:**  
Un sistema puede ser descrito como un conjunto de componentes dentro de un límite que trabajan de manera conjunta para generar una salida a partir de una entrada o entradas específicas. El límite separa al sistema de su entorno, y el sistema interactúa con el entorno mediante señales que atraviesan dicho límite, ya sea en forma de entradas que van del entorno al sistema, o salidas que van del sistema hacia el entorno.

<img src="IMG/sistema.png" alt="sistema" width="400">

**Sistema de instrumentación:**  
Un sistema de instrumentación destinado a realizar mediciones recibe como entrada el valor real de la variable que se mide y genera como salida el valor que ha sido medido.

<img src="IMG/sistema_medicion.png" alt="sistema_medicion" width="400">

**Sensor `(definición 1`):**  
El sensor es el componente del sistema que está en contacto directo con el proceso en el que se mide una variable. Su salida depende de algún modo del valor de dicha variable, y esa salida es utilizada por el resto del sistema de medición para asignarle un valor. Un ejemplo de esto es el termopar, que tiene como entrada la temperatura y como salida una señal eléctrica pequeña.

**Sensor `(definición 2`):**  
Un sensor es un dispositivo que transforma un fenómeno físico en una señal eléctrica. Los sensores son parte de la interfaz entre el mundo físico y el de los dispositivos eléctricos, como las computadoras. La otra parte de esta interfaz está formada por los actuadores, que convierten las señales eléctricas en fenómenos físicos.

**Procesador de señal:**  
Este componente recibe la señal de salida del sensor y la convierte en una forma adecuada para su visualización o para ser transmitida a un sistema de control. Por ejemplo, en el caso de un termopar, un amplificador puede aumentar la fuerza electromotriz (e.m.f.) para que sea lo suficientemente fuerte como para ser registrada en un medidor.

**Función de trasnferencia**

 La función de transferencia describe cómo se relacionan la señal de entrada física y la señal de salida eléctrica. Normalmente, esta relación se ilustra mediante un gráfico que muestra cómo se conectan ambas señales, y los detalles de esta conexión pueden ofrecer una descripción integral de las propiedades del sensor. En el caso de sensores caros que pasan por un proceso de calibración individual, este vínculo puede expresarse mediante una curva de calibración certificada.

 
**Precisión:**

La precisión se refiere a la cercanía de los valores medidos entre sí. asociado a la precisión, la incertidumbre se entiende como el mayor error esperado entre la señal de salida real y la ideal. A veces, esta se presenta como un porcentaje de la salida a escala completa (FSO) o como una fracción de la lectura. Por ejemplo, un termómetro podría garantizar una exactitud dentro del 5% de la salida máxima (FSO). La "precisión" es considerada un término cualitativo en metrología, mientras que la "incertidumbre" es un término cuantitativo. Por ejemplo, un sensor puede tener mejor precisión que otro si su incertidumbre es del 1%, mientras que el otro tiene una incertidumbre del 3%.


**Error:**

El error es la diferencia entre el valor medido y el valor verdadero `(o el valor de referencia`):

$$
\text{Error} = \text{Valor Medido} - \text{Valor Verdadero}
$$

**Error por Histeresis:**

El error por histéresis es la diferencia en la salida del sistema cuando el proceso sigue un ciclo de ida y vuelta, debido a la dependencia de la dirección de la medición.  Existen sensores que no generan el mismo valor de salida cuando el estímulo de entrada se aumenta o disminuye. El rango del error esperado, en relación con la magnitud medida, se conoce como histeresis.

**Sensibilidad:**

 La sensibilidad se describe en función de la relación entre la señal física de entrada y la señal eléctrica de salida. Usualmente, es el cociente entre una pequeña variación en la señal eléctrica y una pequeña variación en la señal física. Esto puede expresarse como la derivada de la función de transferencia respecto a la señal física. Las unidades comunes son voltios/kelvin, milivoltios/kilopascal, entre otras. Un termómetro tendría "alta sensibilidad" si un leve cambio en la temperatura produce un gran cambio en el voltaje.

$$
\text{Sensibilidad} = \frac{\Delta \text{Salida}}{\Delta \text{Entrada}}
$$

**Span o Rango dinámico:**

El rango de señales físicas de entrada que el sensor puede convertir en señales eléctricas se denomina rango dinámico o intervalo. Las señales que se encuentren fuera de este rango se espera que causen una inexactitud demasiado grande. Este intervalo o rango dinámico generalmente lo proporciona el fabricante del sensor, indicando el rango en el que se espera que otras características de rendimiento descritas en las hojas de especificaciones sean válidas. Las unidades comunes son kelvin, pascal, newtons, entre otras.

$$
\text{Span} = \text{Valor Máximo Medible} - \text{Valor Mínimo Medible}
$$

**No Linealidad:**

La máxima desviación de una función de transferencia lineal dentro del rango dinámico especificado. Existen diferentes formas de medir este error, pero la más común es comparar la función de transferencia real con la “mejor línea recta”, que está ubicada en el medio de las dos líneas paralelas que rodean toda la función de transferencia dentro del rango dinámico del dispositivo. Este método es popular porque generalmente hace que los sensores muestren mejores resultados. No obstante, se pueden emplear otras líneas de referencia, por lo que es importante que el usuario asegure que las comparaciones se realicen con la misma referencia. La no linealidad se refiere a la desviación del comportamiento idealmente lineal del sensor. Una medida común de no linealidad es el error máximo entre la salida del sensor y la recta ideal:

$$
\text{No Linealidad} = \frac{\text{Máximo Desvío}}{\text{Span}} \times 100
$$

**Ruido:**

Todos los sensores generan algo de ruido adicional a la señal de salida. En algunos casos, el ruido del sensor es tan bajo que no afecta al sistema, ya que es menor que el ruido del siguiente componente electrónico o las fluctuaciones de la señal física. No obstante, en muchos casos el ruido del sensor limita el desempeño del sistema que lo utiliza. El ruido se distribuye generalmente a través de todo el espectro de frecuencias. Muchas fuentes comunes de ruido producen una distribución de ruido blanco, lo que significa que la densidad espectral de ruido es constante a todas las frecuencias. Un ejemplo claro de esto es el ruido de Johnson en un resistor. Para el ruido blanco, la densidad espectral de ruido se mide en voltios/raíz (Hz). Este tipo de distribución agrega ruido a la medición con una amplitud que depende de la raíz cuadrada del ancho de banda de la medición. Dado que existe una relación inversa entre el ancho de banda y el tiempo de medición, se puede afirmar que el ruido disminuye conforme aumenta el tiempo de medición, siguiendo la raíz cuadrada.

$$
\text{Ruido} = \sigma
$$

donde $\sigma$ es la desviación estándar de la señal de salida.

**Resolución:**

 La resolución de un sensor se refiere a la fluctuación mínima de la señal que se puede detectar. Como las fluctuaciones son fenómenos que ocurren en el tiempo, existe una relación entre la duración de la fluctuación y la amplitud mínima que se puede detectar. Por lo tanto, la definición de resolución debe considerar el tipo de medición que se realiza. Muchos sensores tienen limitaciones debido al ruido con una distribución espectral blanca. En estos casos, la resolución se puede especificar en unidades de señal física/raíz (Hz). La resolución real para una medición específica se obtiene multiplicando esta cantidad por la raíz cuadrada del ancho de banda de medición. Las hojas de especificaciones de los sensores suelen indicar la resolución en unidades de señal/raíz (Hz) o proporcionar la señal mínima detectable para una medición concreta. Si se especifica también la forma de la distribución del ruido, es posible aplicar estos resultados a cualquier tipo de medición. La resolución se refiere a la capacidad del sensor para distinguir cambios pequeños en la variable medida. Generalmente, la resolución es el menor cambio detectable, y se puede expresar como:

$$
\text{Resolución} = \frac{\text{Span}}{2^N}
$$

donde \(N\) es el número de bits de la conversión analógica a digital (ADC).

**Ancho de Banda:**

Todos los sensores presentan tiempos de respuesta limitados ante un cambio instantáneo en la señal física. Además, muchos sensores tienen tiempos de decaimiento, los cuales indican el tiempo que le toma al sensor recuperar su valor original tras un cambio abrupto en la señal física. Los recíprocos de estos tiempos se corresponden con las frecuencias de corte superior e inferior. El ancho de banda de un sensor es el intervalo de frecuencias comprendido entre estas dos frecuencias de corte. El ancho de banda es la frecuencia máxima a la que el sistema puede responder sin atenuación significativa de la señal. Generalmente se expresa en hertzios (Hz):

$$
\text{Ancho de Banda} = \text{Frecuencia máxima de respuesta}
$$

  <h3>Ejemplo</h3>
  Para dar más sustancia a estas definiciones, identificaremos los valores numéricos de estos parámetros para un acelerómetro comercial, el ADXL150 de Analog Devices.

  $$ V(\text{Acc}) = 1.5V + (\text{Acc} \times 167 \, \text{mV/g}) $$  
    
  Esta expresión puede usarse para predecir el comportamiento del sensor y contiene información sobre la sensibilidad y el desplazamiento en la salida del sensor.

```python
import numpy as np
import plotly.graph_objects as go

# Función de transferencia con sensibilidad ajustada por temperatura
def calcular_voltaje(acc, temp=20):
    sensibilidad_nominal = 0.167e-3  # 167 mV/g
    delta_temp = temp - 20  # Cambio de temperatura respecto a 20°C
    sensibilidad_temperatura = sensibilidad_nominal * (1 + 0.00025 * delta_temp)  # Ajuste de sensibilidad
    return 1.5 + (acc * sensibilidad_temperatura)  # Calcular el voltaje

# Crear valores de aceleración
acc_values = np.linspace(-650, 650, 1000)  # Aceleración desde -650g a 650g

# Simular ruido con distribución normal (media = 0, desviación estándar = 5)
ruido = np.random.normal(0, 5, acc_values.shape)

# Aplicar el ruido a los valores de aceleración
acc_values_ruido = acc_values + ruido

# Calcular los voltajes correspondientes con aceleración con ruido
volt_values = calcular_voltaje(acc_values_ruido)

# Líneas para representar la desviación máxima de linealidad (±0.2% de la salida máxima)
max_voltaje = 3.5
deviation = max_voltaje * 0.002  # 0.2% de 3.5V

# La línea ideal de la función de transferencia sin ruido
volt_lineal_ideal = calcular_voltaje(acc_values)

# Líneas de desviación de linealidad (±0.2%)
volt_lineal_ideal_plus = volt_lineal_ideal + deviation
volt_lineal_ideal_minus = volt_lineal_ideal - deviation

# Crear la figura interactiva
fig = go.Figure()

# Agregar la gráfica de la función de transferencia sin ruido (azul discontinua)
fig.add_trace(go.Scatter(x=acc_values, y=volt_lineal_ideal,
                         mode='lines', name='V = 1.5 + Acc × 0.167e-3 V/g (sin ruido)', 
                         line=dict(dash='dash', color='blue', width=4)))  # Hice la línea más gruesa

# Agregar la gráfica con ruido (roja)
fig.add_trace(go.Scatter(x=acc_values, y=volt_values,
                         mode='lines', name='V con Ruido', line=dict(color='red')))

# Marcar el rango dinámico garantizado de ±2g con líneas verticales
fig.add_vline(x=-2, line=dict(color="green", width=2), name="Rango Garantizado (-2g a 2g)")
fig.add_vline(x=2, line=dict(color="green", width=2))

# Líneas de desviación de linealidad (±0.2%)
fig.add_trace(go.Scatter(x=acc_values, y=volt_lineal_ideal_plus,
                         mode='lines', name='+0.2% Linealidad', line=dict(dash='dot', color='purple')))
fig.add_trace(go.Scatter(x=acc_values, y=volt_lineal_ideal_minus,
                         mode='lines', name='-0.2% Linealidad', line=dict(dash='dot', color='purple')))

# Considerar diferentes rangos de temperatura y ajustar la salida del voltaje
temperaturas = np.linspace(-10, 50, 5)  # Rango de temperaturas entre -10°C y 50°C
for temp in temperaturas:
    volt_temp = calcular_voltaje(acc_values, temp)  # Calcular voltaje para cada temperatura
    fig.add_trace(go.Scatter(x=acc_values, y=volt_temp,
                             mode='lines', name=f'Temperatura: {temp}°C', 
                             line=dict(dash='solid', width=2)))  # Cerré el paréntesis aquí

# Configurar el diseño de la gráfica
fig.update_layout(
    title="Función de Transferencia: Aceleración vs. Voltaje con Ruido, Efecto de Linealidad y Temperatura",
    xaxis_title="Aceleración (g)",
    yaxis_title="Voltaje (V)",
    legend_title="Leyenda",
    showlegend=True,
    template="plotly_dark",
    xaxis=dict(range=[-650, 650]),
    yaxis=dict(range=[1.0, 3.5]),
    updatemenus=[dict(
        type="buttons",
        showactive=False,
        buttons=[dict(
            label="Restablecer Zoom",
            method="relayout",
            args=[{"xaxis.range": [-650, 650], "yaxis.range": [1.0, 3.5]}]
        )]
    )]
)

# Mostrar la gráfica interactiva
fig.show()

```

### Aceleración vs. Voltaje con Ruido, Efecto de Linealidad y Temperatura  

1. **Función de Transferencia (Sin Ruido)**  

    La línea azul discontinua representa la función de transferencia nominal del sensor sin ningún tipo de ruido.  
    Esta es la relación ideal entre la aceleración (en g) y el voltaje de salida (en V), dada por la ecuación:  

    $ V = 1.5 + (Acc \times 0.167 \times 10^{-3}) $  

    Donde:  
    - $ V $ es el voltaje de salida (en V).  
    - $ Acc $ es la aceleración (en g).  
    - $ 1.5V $ es el voltaje base cuando no hay aceleración ($ Acc = 0 $).  
    - La sensibilidad del sensor es $ 167 $ mV/g $ = 0.167 \times 10^{-3} $ V/g.  

2. **Voltaje con Ruido**  

    La línea roja muestra el voltaje calculado con ruido.  
    El ruido se simula utilizando una distribución normal con una desviación estándar de **5**, lo que introduce fluctuaciones en los valores de aceleración, afectando el voltaje de salida.  
    Esta representación nos permite ver cómo las mediciones pueden variar debido a interferencias externas o incertidumbres en el sistema.  

3. **Rango Dinámico Garantizado**  

    Las líneas verdes verticales marcan el **rango dinámico garantizado** del sensor, que es de **±2g**.  
    Este es el rango de aceleración dentro del cual el sensor proporciona mediciones confiables.  
    Fuera de este rango, las mediciones podrían volverse inexactas o no lineales.  

4. **Efecto de Linealidad (±0.2%)**  

    Las líneas moradas discontinuas representan el **rango de desviación de linealidad**, es decir, las variaciones máximas permitidas respecto a la línea ideal.  
    En este caso, la **linealidad del sensor está garantizada dentro de un 0.2% de la salida máxima**.  
    Esto significa que las mediciones de voltaje pueden desviarse de la función ideal hasta un **0.2%** debido a la naturaleza no perfecta del sensor.  

    La ecuación para calcular la desviación de linealidad es:  

    Desviación Linealidad = $\pm (0.002 \times V_{max})$  

    En este caso, el voltaje máximo es **3.5V**, por lo que la desviación es:  

    Desviación = $3.5V \times 0.002 = 0.007V = 7$  

    Por lo tanto, las líneas moradas representan un desplazamiento de **±7 mV** respecto a la función ideal de transferencia.  

5. **Ajuste de Sensibilidad por Temperatura**  

    Las líneas adicionales con diferentes colores muestran cómo el voltaje cambia con la **temperatura**.  
    El ajuste de sensibilidad por temperatura se calcula considerando un cambio de **±0.025% por °C** respecto a la temperatura nominal de **20°C**.  

    La ecuación de sensibilidad ajustada por temperatura es:  

    Sensibilidad Ajustada =  Sensibilidad Nominal $\times (1 + (0.00025 \times \Delta T)) $  

    Donde:  
    - **Sensibilidad Nominal** = $0.167 \times 10^{-3}$ V/g (sensibilidad a temperatura ambiente de **20°C**).  
    - **$\Delta T$** es la diferencia de temperatura con respecto a **20°C**.  

    Esto implica que por cada **grado Celsius** que la temperatura se desvíe de **20°C**, la sensibilidad del sensor se ajustará en un **0.025%** de su valor nominal.  
    Las líneas sólidas en la gráfica representan este ajuste de sensibilidad para distintas temperaturas.  

    **Ejemplo**:  
    Si la temperatura aumenta a **30°C** ($\Delta T = 10°C$), la sensibilidad ajustada será:  

    Sensibilidad Ajustada = $0.167 \times 10^{-3} \times (1 + (0.00025 \times 10))$  

    $=0.167 \times 10^{-3} \times 1.0025 = 0.167417 \times 10^{-3}$ V/g  

6. **Botón de Restablecer Zoom**  

    El gráfico también incluye un botón interactivo **"Restablecer Zoom"**, que permite al usuario restablecer fácilmente los rangos de los ejes **x (aceleración)** y **y (voltaje)** a los valores predeterminados.

### Norma para Simbología P&ID

La simbología utilizada en los diagramas de tuberías e instrumentación (P&ID) se puede consultar en varias normas, dependiendo de la región y el ámbito industrial. A continuación, se mencionan algunas de las normas más comunes:

1. **ISA-5.1**  
   - **Título**: "Instrumentation Symbols and Identification"  
   Esta es la norma más ampliamente utilizada para la simbología de instrumentos y señales en los diagramas de control, incluidos los P&ID. Define símbolos estándar para representar equipos, líneas de proceso, instrumentos y componentes del sistema de control.  
   - Para más información, puedes consultar la norma completa [aquí](https://integrated.cc/cse/Instrumentation_Symbols_and_Identification.pdf).

2. **ANSI/ISA-5.1**  
   - Similar a la ISA-5.1, pero adaptada al contexto estadounidense. Se utiliza en América del Norte y está enfocada en estándares específicos de esa región.

3. **ISO 10628**  
   - **Título**: "Diagrams for the Chemical and Process Industries"  
   Esta norma proporciona directrices internacionales para diagramas de flujo de procesos, incluidas las representaciones de instrumentos y equipos en los diagramas de proceso e instrumentación. Es de uso común en la industria química y de procesos.

4. **IEC 81346**  
   - **Título**: "Industrial Systems, Installations, and Equipment and Industrial Products"  
   Esta norma aborda la identificación y clasificación de los equipos y sistemas industriales, incluyendo la simbología para diagramas eléctricos y de instrumentación. Es útil para empresas que trabajan en proyectos internacionales e industriales a gran escala.

  <h4>Código Arduino para leer un potenciómetro y enviar el voltaje por serial</h4>
  <div style="background-color: #f4f4f4; border: 2px solid #8e44ad; padding: 15px; border-radius: 5px;">
    <pre><code style="color: #333333; font-family: Consolas, monospace;">
const int potPin = A0;  // Pin donde está conectado el potenciómetro

void setup() {
  Serial.begin(9600);  // Inicializa la comunicación serial a 9600 baudios
}

void loop() {
  int sensorValue = analogRead(potPin);  // Lee el valor del potenciómetro
  float voltage = sensorValue * (5.0 / 1023.0);  // Convierte el valor a voltaje
  
  Serial.println(voltage);  // Envía el voltaje por el puerto serial

  delay(500);  // Espera 500 ms antes de la siguiente lectura
}
    </code></pre>
  </div>

  <h4>Código Python para almacenar los datos y graficar el voltaje vs muestra</h4>
  <p>El siguiente código en Python lee los valores de voltaje enviados por el Arduino a través del puerto serial, los almacena en un arreglo y luego los grafica:</p>

  <div style="background-color: #f4f4f4; border: 2px solid #8e44ad; padding: 15px; border-radius: 5px;">
    <pre><code style="color: #333333; font-family: Consolas, monospace;">
import serial
import matplotlib.pyplot as plt
import numpy as np

ser = serial.Serial('COM3', 9600)  # Cambia 'COM3' por el puerto correcto

voltajes = []
muestras = []

num_muestras = 100  

for i in range(num_muestras):
    line = ser.readline().decode('utf-8').strip()  
    try:
        voltaje = float(line)  # Convierte el valor de la línea a un número flotante
        voltajes.append(voltaje)  # Agrega el voltaje al arreglo
        muestras.append(i)  # Agrega el índice de la muestra
    except ValueError:
        pass 

plt.plot(muestras, voltajes, label='Voltaje', color='b')
plt.xlabel('Muestras')
plt.ylabel('Voltaje (V)')
plt.title('Voltaje vs Muestras')
plt.legend()
plt.grid(True)
plt.show()
    </code></pre>
  </div>

  <h4>Explicación del Código Python</h4>
  <p>El código Python realiza las siguientes tareas:</p>
  <ul>
    <li><b>Configura el puerto serial:</b> Se establece la comunicación con el Arduino a través del puerto serial. Asegúrate de cambiar `'COM3'` por el puerto correcto de tu sistema.</li>
    <li><b>Lectura de datos del Arduino:</b> El código lee los valores de voltaje enviados por el Arduino a través del puerto serial, los almacena en dos listas: una para los voltajes y otra para las muestras (índices de lectura).</li>
    <li><b>Almacenamiento de datos:</b> Los valores de voltaje se convierten a flotantes y se agregan a una lista para su posterior análisis.</li>
    <li><b>Generación de la gráfica:</b> Se utiliza la librería `matplotlib` para graficar los voltajes en función del número de muestra. La gráfica permite visualizar cómo varían los voltajes a lo largo del tiempo.</li>
  </ul>
